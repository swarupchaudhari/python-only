```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Universe of discourse
x = np.linspace(0, 100, 500)
```

```python
# 1. Triangular Membership Function
# -----------------------------
def triangular(x, a, b, c):
    return np.maximum(np.minimum((x-a)/(b-a), (c-x)/(c-b)), 0)
```

```python
# 2. Trapezoidal Membership Function
# -----------------------------
def trapezoidal(x, a, b, c, d):
    return np.maximum(np.minimum(np.minimum((x-a)/(b-a), 1), (d-x)/(d-c)), 0)
```

```python
# 3. Gaussian Membership Function
# -----------------------------
def gaussian(x, mean, sigma):
    return np.exp(-((x-mean)**2)/(2*sigma**2))
```

```python
# 4. Generalized Bell Membership Function
# -----------------------------
def bell(x, a, b, c):
    return 1 / (1 + np.abs((x-c)/a)**(2*b))
```

```python
# 5. Sigmoid Membership Function
# -----------------------------
def sigmoid(x, a, c):
    return 1 / (1 + np.exp(-a*(x-c)))
```

```python
# Calculate Membership Values
# -----------------------------
tri = triangular(x, 10, 30, 50)
trap = trapezoidal(x, 20, 35, 65, 80)
gauss = gaussian(x, 50, 10)
bellmf = bell(x, 15, 2, 50)
sig = sigmoid(x, 0.2, 50)
```

```python
# Plotting
# -----------------------------
plt.figure(figsize=(10,6))

plt.plot(x, tri, label='Triangular')
plt.plot(x, trap, label='Trapezoidal')
plt.plot(x, gauss, label='Gaussian')
plt.plot(x, bellmf, label='Bell')
plt.plot(x, sig, label='Sigmoid')

plt.title("Different Fuzzy Membership Functions")
plt.xlabel("Universe of Discourse")
plt.ylabel("Membership Value")
plt.grid(True)
plt.legend()
plt.show()
```
